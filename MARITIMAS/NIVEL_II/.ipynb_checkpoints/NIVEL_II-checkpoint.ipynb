# Enunciado del ejemplo:

<i>Empleando el ajuste de Negro y Varela, se pretende obtener el valor de la probabilidad de fallo y punto de diseño de una obra de un dique de tipología rompeolas con piezas cúbicas de hormigón en masa para protección de su manto principal, dimensión del bloque de Dn<sub>50</sub> = 2,50 m y peso específico del hormigón de  γ= 2,40 T/m 3 , considerando estos dos
parámetros (Dn<sub>50</sub> y γ) como variables deterministas, es decir, de valor fijo o no aleatorias.</i>

**Calculamos:**

$\displaystyle \Delta=\frac{\gamma}{\gamma_{w}}-1 $

```python
gamma= 2.4
gamma_w=1.03
Dn=2.5
Delta=(gamma/gamma_w)-1
```

```python
Delta
```

```python
Delta*Dn
```

**Lo primero que necesitamos es definir la función de fallo Z que relaciona los términos de resistencia (R) y de solicitación (S). Cuando R > S el sistema es seguro y Z > 0. Cuando R < S el sistema falla y Z < 0. Cuando R = S estamos en el caso
de estado límite con Z = 0.**

$$\displaystyle Z=R-S $$

**En este caso concreto, definiremos los conceptos de resistencia y solicitación en términos de altura de ola significante.
Como término resistente tendremos la altura de ola resistida por el bloque de hormigón de dimensiones y peso específico
conocidos (variables deterministas). Como término de solicitación tendremos el valor de la altura de ola significante
medida en una boya situada en aguas profundas, asociada a un valor de periodo de retorno T<sub>r</sub> y propagada hasta el punto
de diseño del dique.**

$$\displaystyle Z = N_{s} \cdot \Delta \cdot D_{n50} - (Hs)_{Tr}$$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.stats import multivariate_normal

# Definir medias y desviaciones estándar
mean1, std1 = 1.97, 0.125
mean2, std2 = 5, 1.20

# Crear malla de valores para X1 y X2
x1 = np.linspace(mean1 - 3*std1, mean1 + 3*std1, 500)
x2 = np.linspace(mean2 - 3*std2, mean2 + 3*std2, 500)
X1, X2 = np.meshgrid(x1, x2)
pos = np.empty(X1.shape + (2,))
pos[:, :, 0] = X1
pos[:, :, 1] = X2

# Calcular la matriz de covarianza
cov_matrix = np.array([[std1**2, 0], [0, std2**2]])

# Calcular la función de densidad de probabilidad multivariada
rv = multivariate_normal([mean1, mean2], cov_matrix)
pdf_values = rv.pdf(pos)

# Crear el gráfico tridimensional
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X1, X2, pdf_values, cmap='viridis')

# Configuración del gráfico
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('f(X1, X2)')
ax.set_title('Distribución Conjunta Gaussiana')



# Mostrar el gráfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# Definir medias y desviaciones estándar
mean1, std1 = 1.97, 0.125
mean2, std2 = 5, 1.20

# Crear malla de valores para X1 y X2
x1 = np.linspace(0, 4, 100)
x2 = np.linspace(0, 10, 100)
X1, X2 = np.meshgrid(x1, x2)
pos = np.empty(X1.shape + (2,))
pos[:, :, 0] = X1
pos[:, :, 1] = X2

# Calcular la matriz de covarianza
cov_matrix = np.array([[std1**2, 0], [0, std2**2]])

# Calcular la función de densidad de probabilidad multivariada
rv = multivariate_normal([mean1, mean2], cov_matrix)
pdf_values = rv.pdf(pos)

# Crear el gráfico tridimensional
fig = plt.figure(figsize=(12, 5))

# Subgráfico 3D
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X1, X2, pdf_values, cmap='viridis', alpha=0.8)
ax1.set_xlabel('X1')
ax1.set_ylabel('X2')
ax1.set_zlabel('f(X1, X2)')
ax1.set_title('Distribución Conjunta 3D')

# Ajustar la vista para girar en sentido contrario a las agujas del reloj
ax1.view_init(elev=30, azim=135)

# Añadir el corte producido por el plano perpendicular
x1_cut = np.linspace(0, 4, 100)
x2_cut = 3.325 * x1_cut
ax1.plot(x1_cut, x2_cut, np.zeros_like(x1_cut), color='red', label='Corte 3.325*X1-X2=0')
ax1.legend()

# Subgráficos de las proyecciones
ax2 = fig.add_subplot(122)
ax2.contour(X1, X2, pdf_values, cmap='viridis', levels=20)
ax2.plot(x1_cut, x2_cut, color='red', label='Corte 3.325*X1-X2=0')
ax2.set_xlabel('X1')
ax2.set_ylabel('X2')
ax2.set_xlim([0, 4])
ax2.set_ylim([0, 10])
ax2.set_title('Proyección en el Plano X1-X2')
ax2.legend()

plt.tight_layout()
plt.show()
```
